```python
import torch
import torchvision
def laplace_filter(img):
    batchsize,channels,H, W = img.shape
    print(batchsize,channels,H, W)
    lp_img = torch.zeros(img.shape)
    K = torch.tensor([[0., 1., 0.],[1., -4., 1.], [0., 1., 0.]])
    for n_batch in range(0,batchsize):
        for n_channels in range(0,channels):
            for i in range(1,H-1):
                for j in range(1,W-1):
                    lp_img[n_batch,n_channels,i,j] = (img[n_batch,n_channels,i-1 : i+2, j-1 : j+2] * K).sum() 
    out = lp_img[:,:,1:-1, 1:-1]
    return out
```

```python
import torch
import cv2
import numpy as np
torch.set_printoptions(threshold=np.inf)
# 读取图片
path = 'C:/Users/17865/Desktop/laplace/input/'


file_in = path + 'input1.png' 
img = cv2.imread(file_in)
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img_tensor = torch.tensor(img)
img_tensor = torch.unsqueeze(img_tensor, 0)
img_tensor = torch.unsqueeze(img_tensor, 0)  #torch.Size([1, 1, 339, 141])
b = laplace_filter(img_tensor)
print(b)
b = torch.squeeze(b,0)
b = torch.squeeze(b,0)
b = b.numpy()
cv2.imwrite("b.jpg", b)
```
